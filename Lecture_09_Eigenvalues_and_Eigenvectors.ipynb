```python
import numpy as np
```

```python
a=np.array([[2,-2,3],[1,1,1],[1,3,-1]])
```

```python
print(a)
```

```python
#eigen pair
```

```python
values,vector=np.linalg.eig(a)
```

```python
print(values)
```

```python
print(vector)
```

```python
# these eigen vector are of unit lenght
```
