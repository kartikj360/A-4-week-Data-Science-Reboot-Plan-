```python
import numpy as np
```

```python
p=np.array([[1,7],[1,8],[2,8]])
print(p)
```

```python
q=np.array([[7,1],[8,1],[1,5]])
print(q)
```

```python
c=np.add(p,q)
print(c)
```

```python
c2=np.subtract(q,p)
```

```python
print(c2)
```

```python
c3=np.dot(p,q)
print(c3)
```

```python
det=np.linalg.det(c3)
print(det)
```

```python
invs=np.linalg.inv(c3)
```
